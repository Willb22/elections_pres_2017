#### Pour visualiser des données géographiques sur les pourcentages d'absention par bureaux de votes, nous devons procéder à la fusion de deux bases de données
#### PR17_BVot_T1_FE (copy).txt contient les informations électorales par bureaux de vote
#### geo_bureaux_de_vote.csv contient les  adresses, et latitudes, longitudes géographiques des bureaux de vote
#### Nous allons voir, pour l'exemple du département de Paris (75) comment faire correspondre les codes bureaux de votes tels que précisés dans une base de données, par rapport à l'autre


### Paris, 18eme circonscription
#### 4 bureaux sont dans le 75009, 40 dans le 75018
### Dans df_paris (non-geo), 18eme circonscription
#### les bureaux du 75009 ont pour code 9***
#### les bureaux du 75018 ont pour code 18***
### Dans geo_paris, 18eme circonscription
#### les codes de bureaux de vote sont les suffixes des codes indiqués dans df_paris, car le code postal est précisé ailleurs. un zero est parfois intercallé

```python
df = pd.read_csv("../data/raw/PR17_BVot_T1_FE (copy).txt", encoding = "ISO-8859-1", sep =';', decimal =',')

df
```

```python
df_paris = df[df['Libellé du département']=='Paris']
#df_paris['Code de la circonscription'] = df_paris['Code de la circonscription'].apply(lambda x : str(x))
#df_paris['Code du b.vote'] = df_paris['Code du b.vote'].apply(lambda x : str(x))
#df_paris
```

Il arrive parfois que la circonscription corresponde à l'arrondissement.  
C'est le cas pour certains bureaux de votes dans le premier et 18ème arrondissement.  
En quittant les deux derniers charactères dans 'Code du b.vote', nous souhaitons voir si cela nous donne l'arrondissment.
Ci-dessous, nous voyons les 50 bureaux de votes pour lequel la circonscription correspond à l'arrondissement

```python
df_paris[df_paris['Code de la circonscription'].apply(lambda x : str(x)) == df_paris['Code du b.vote'].apply(lambda x : str(x)).apply(lambda x : x[:-2])] #les premiers caractères du code bureaux de vote correspondent à l'arrondissement et pas forcément la valeur de la cirsconscription
```

Ensuite nous observons de plus près pour les 846 autres bureaux de vote dans Paris, lorsque quitter les deux premier caractères dans 'Code du b.vote' ne correspond pas à 'Code de la circonscription'.  
On peut identifier les arrondissments que contiennent chaque circonscription de Paris

```python
differe = df_paris[df_paris['Code de la circonscription'].apply(lambda x : str(x)) != df_paris['Code du b.vote'].apply(lambda x : str(x)).apply(lambda x : x[:-2])] #les premiers caractères du code bureaux de vote correspondent à l'arrondissement et pas forcément la valeur de la cirsconscription
differe
```

Nous vérifions si en quittant les deux derniers caractères de 'Code du b.vote', nous retrouvons le découpage des circonscriptions dans Paris en 2017, selon les arrondissements

```python
for i in df_paris['Code de la circonscription'].unique():

    circonscription = df_paris[df_paris['Code de la circonscription']==i]

    arrondissements = circonscription['Code du b.vote'].apply(lambda x : str(x)).apply(lambda x : x[:-2]).unique()
    
    print("Circonscription {} contient les arrondissements {}".format(i, arrondissements))
    
    
```

Nous avons donc les informations nous permettant de recouper avec les bureaux de votes précisés dans la base de données géographique geo_bureaux_de_vote.csv  
On procède maintenant à la création d'une colonne qui permettra la fusion avec les infos géographiques contenues dans geo_bureaux_de_vote.csv 

```python
df_paris['Code du département'].apply(lambda x : str(x)) + '-' +df_paris['Code de la circonscription'].apply(lambda x : str(x)).apply(lambda x : '0'+x if len(x)<2 else x) #harmonize with geo_data
```

```python
df_paris['Code du b.vote'].apply(lambda x : str(x)).apply(lambda x : x[:-2]).unique()
```

#### Format de la colonne pour faire une jointure avec la base géographique  
75- + code de la circonscription + _ + arrondissemnet + code bureaux de vote  
Cette information précise devrait éviter d'associer des bureaux de votes qui ne correspondent pas

```python
df_merge_col = df_paris['Code du département'].apply(lambda x : str(x)) + '-' +df_paris['Code de la circonscription'].apply(lambda x : str(x)).apply(lambda x : '0'+x if len(x)<2 else x) + '_' + df_paris['Code du b.vote'].apply(lambda x : str(x)).apply(lambda x : '0'+x if len(x)<4 else x)
df_merge_col
```

```python
df_paris['col_merge'] = df_merge_col #creation de la colonne pour faire la jointure
df_paris
```

## Base de données geographiques

```python
geo = pd.read_csv('../data/raw/geo_bureaux_de_vote.csv')
postal = geo['code_postal'].fillna('00')
boo = postal.apply(lambda x : True if x.startswith('75') else False)
geo_paris = geo[boo]
geo_paris
```

```python
print("On constate {} bureaux de votes dans la base géographique, et {} dans df_paris".format(geo_paris.shape[0], df_paris.shape[0]))
```

```python
geo_paris['circonscription_code'].unique() #harmonisé par rapport au format que l'on a crée dans df_paris
```

#### Format de la colonne pour faire une jointure
75- + code de la circonscription + _ + arrondissement + code bureaux de vote  
Cette information précise devrait éviter d'associer des bureaux de votes qui ne correspondent pas

```python
geo_col_merge = geo_paris['circonscription_code']+ '_' + geo_paris['code_postal'].apply(lambda x : str(x)[-2:]) + geo_paris['code'].apply(lambda x : str(x)[-2:])  
#geo_col_merge = list(geo_col_merge)
geo_col_merge
```

```python
geo_paris['col_merge'] = geo_col_merge
geo_paris= geo_paris[['longitude', 'latitude','geo_adresse', 'col_merge']]
geo_paris
```

## Jointure des deux bases

```python
df_merged = pd.merge(geo_paris, df_paris, left_on='col_merge', right_on='col_merge')
df_merged
```

```python
df_merged.to_csv('./geo_75.csv')
```
