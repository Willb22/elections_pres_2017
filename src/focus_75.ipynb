### Pour visualiser des données géographiques sur les pourcentages d'absention par bureaux de votes, nous devons procéder à la fusion de deux bases de données
### PR17_BVot_T1_FE (copy).txt contient les informations électorales par bureaux de vote
### geo_bureaux_de_vote.csv contient les  adresses, et latitudes, longitudes géographiques des bureaux de vote
### Nous allons voir, pour l'exemple du département de Paris (75) comment faire correspondre les codes bureaux de votes tels que précisés dans une base de données, par rapport à l'autre


### Paris, 18eme circonscription
#### 4 bureaux sont dans le 75009, 40 dans le 75018
### Dans df_paris (non-geo), 18eme circonscription
#### les bureaux du 75009 ont pour code 9***
#### les bureaux du 75018 ont pour code 18***
### Dans geo_paris, 18eme circonscription
#### les codes de bureaux de vote sont les suffixes des codes indiqués dans df_paris, car le code postal est précisé ailleurs. un zero est parfois intercallé

```python
df = pd.read_csv("../data/raw/PR17_BVot_T1_FE (copy).txt", encoding = "ISO-8859-1", sep =';', decimal =',')

df
```

```python
geo = pd.read_csv('../data/raw/geo_bureaux_de_vote.csv')
geo.shape
```

```python
df_paris = df[df['Libellé du département']=='Paris']
#df_paris
```

```python
df_paris[df_paris['Code de la circonscription'].apply(lambda x : str(x)) != df_paris['Code du b.vote'].apply(lambda x : str(x)).apply(lambda x : x[:-2])] #les premiers caractères du code bureaux de vote correspondent à l'arrondissement et pas forcément la valeur de la cirsconscription
```

```python
df_paris['Code du département'].apply(lambda x : str(x)) + '-' +df_paris['Code de la circonscription'].apply(lambda x : str(x)).apply(lambda x : '0'+x if len(x)<2 else x) #harmonize with geo_data
```

```python
df_merge_col = df_paris['Code du département'].apply(lambda x : str(x)) + '-' +df_paris['Code de la circonscription'].apply(lambda x : str(x)).apply(lambda x : '0'+x if len(x)<2 else x) + '_' + df_paris['Code du b.vote'].apply(lambda x : str(x)).apply(lambda x : '0'+x if len(x)<4 else x)
df_merge_col
```

```python
df_paris['col_merge'] = df_merge_col #creation de la colonne pour faire la jointure
df_paris
```
