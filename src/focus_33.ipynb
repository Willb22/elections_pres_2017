```python
duplicates = list()
for circon in geo_33['circonscription_code'].unique():
    #print(circon)
    geo_cir = geo_33[geo_33['circonscription_code']==circon]
    if len(geo_cir['code'].unique()) != len(geo_cir):
        print(circon, " nombre de codes bureaux ", len(geo_cir['code'].unique()), " nombre de bureaux ",len(geo_cir) )
        duplicates.append(abs(len(geo_cir['code'].unique())-len(geo_cir)))

print('total duplicates ', sum(duplicates))
```

Nous devons homogénéiser la longueur du code bureaux de vote dans geo_33

```python
df_merge = df_33['Code de la commune'].apply(lambda x : str(x + 33000)) +'_' + df_33['Code du b.vote']
df_33[df_merge.duplicated()]
```

```python
df_33['merge'] = df_merge
df_33
```

#### Creation de la colonne de fusion pour la base de données géographique

```python
geo_merge = geo_33['commune_code']+'_'+geo_33['code'].apply(lambda x : '0'+x if len(x) ==3 else '00'+x if len(x)==2 else '000'+x if len(x) ==1 else x)
geo_33[geo_merge.duplicated()]
```

```python
geo_33['merge'] = geo_merge
geo_33
```

```python
geo_to_merge = geo_33[['merge', 'latitude', 'longitude', 'geo_adresse']]
geo_to_merge
```
